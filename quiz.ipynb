10.1) Напишите список линтеров которые используются для проверки кода и дайте их краткую характеристику.
-Pylint. Проверяет код на наличие ошибок и на соблюдение стандартов (длина строк, названия переменных и т. п.).
-flake8. Проверяет код на соответствие стандартам PEP 8, выявляет синтаксические ошибки и предупреждения, которые могут повлиять на выполнение программы.
-Black. Автоматически форматирует код в соответствии с руководством по стилю PEP 8. В отличие от традиционных линтеров, которые только подчёркивают проблемы стиля, Black переформатирует код на месте, внеся отступы, длину строки и другие изменения стиля. 
-mypy. Проверяет, соответствуют ли значения переменных их объявленным типам, и помогает обнаруживать ошибки до запуска кода.
